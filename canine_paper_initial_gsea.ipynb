```python
pemphigus = pd.read_csv('GSE77361/gsea/msigdb_hallmarks.gmt/gseapy.gsea.gene_set.report.csv', index_col=0)
early_ctcl = pd.read_csv('GSE113113/gsea/early/msigdb_hallmarks.gmt/gseapy.gsea.gene_set.report.csv', index_col=0)
late_ctcl = pd.read_csv('GSE113113/gsea/late/msigdb_hallmarks.gmt/gseapy.gsea.gene_set.report.csv', index_col=0)
dle = pd.read_csv('GSE81071/gsea/msigdb_hallmarks.gmt/gseapy.gsea.gene_set.report.csv', index_col=0)
```

```python
dott_plot.dot_plot([pemphigus, early_ctcl, late_ctcl, dle],
                   names=['Pemphigus', 'Early CTCL', 'Late CTCL', 'DLE'])
```

```python
dott_plot.dot_plot([early_ctcl, late_ctcl, dle],
                   names=['Early CTCL', 'Late CTCL', 'DLE'])
```

```python
dle.columns =['DLE' if x=='fdr' else x for x in dle.columns]
early_ctcl.columns =['Early CTCL' if x=='fdr' else x for x in early_ctcl.columns]
late_ctcl.columns =['Late CTCL' if x=='fdr' else x for x in late_ctcl.columns]
pemphigus.columns =['Pemphigus' if x=='fdr' else x for x in pemphigus.columns]
```

```python
df = pd.concat([pemphigus['Pemphigus'], dle['DLE'], early_ctcl['Early CTCL'], late_ctcl['Late CTCL']], axis=1)
```

```python
heatmap.sns_clustermap(df)
```
